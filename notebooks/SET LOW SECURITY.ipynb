```sql
create table if not exists metacatalog.goldschema.product_map (
  email STRING,
  category STRING
)
```

```sql
CREATE OR REPLACE FUNCTION metacatalog.goldschema.row_level_security(p_category STRING)
RETURNS BOOLEAN
LANGUAGE SQL
RETURN
(
exists(
select * from metacatalog.goldschema.product_map
where email = current_user()
and category = p_category)
)
```

```sql
alter table metacatalog.goldschema.dim_product
SET ROW FILTER metacatalog.goldschema.row_level_security on (Category)
```
